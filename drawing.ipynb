```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
data_dir_1 = 'D:\Code\python\dog breed\\resnet34_test1'
data_dir_2 = 'D:\Code\python\dog breed\\resnet34_test2'
nums_1 = 7
nums_2 = 7
loss_1 = []
valid_loss_1 = []
loss_2 = []
valid_loss_2 = []

for i in range(nums_1):
    fname = 'loss_resnet34_no_pretrained_' + str(i) + '.csv'
    path = os.path.join(data_dir_1, fname)
    df = pd.read_csv(path, header=None, skiprows=1)
    df.columns = ['index', 'value']
    loss_1 += list(df['value'])

for i in range(nums_1):
    fname = 'valid_los_resnet34_no_pretrained_' + str(i) + '.csv'
    path = os.path.join(data_dir_1, fname)
    df = pd.read_csv(path, header=None, skiprows=1)
    df.columns = ['index', 'value']
    valid_loss_1 += list(df['value'])

for i in range(nums_2):
    fname = 'loss_resnet34_no_pretrained_' + str(i) + '.csv'
    path = os.path.join(data_dir_2, fname)
    df = pd.read_csv(path, header=None, skiprows=1)
    df.columns = ['index', 'value']
    loss_2 += list(df['value'])

for i in range(nums_2):
    fname = 'valid_los_resnet34_no_pretrained_' + str(i) + '.csv'
    path = os.path.join(data_dir_2, fname)
    df = pd.read_csv(path, header=None, skiprows=1)
    df.columns = ['index', 'value']
    valid_loss_2 += list(df['value'])

x_loss_1 = np.linspace(1, nums_1 * 20, num=len(loss_1))
x_valid_loss_1 = np.linspace(1, nums_1 * 20, num=len(valid_loss_1))
x_loss_2 = np.linspace(1, nums_2 * 20, num=len(loss_2))
x_valid_loss_2 = np.linspace(1, nums_2 * 20, num=len(valid_loss_2))

plt.plot(x_loss_1, loss_1, color='#1f77b4', label='Unoptimized Training Loss')       # 深蓝色
plt.plot(x_valid_loss_1, valid_loss_1, color='#a3aed6', label='Unoptimized Validation')  # 浅蓝色

plt.plot(x_loss_2, loss_2, color='#2ca02c', label='Optimized Training')       # 深绿色
plt.plot(x_valid_loss_2, valid_loss_2, color='#98df8a', label='Optimized Validation')  # 浅绿色

plt.xlabel('Epochs')
plt.ylabel('Loss Value')
plt.title('Training and Validation Loss')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.grid(True)
plt.legend()
plt.show()
```
