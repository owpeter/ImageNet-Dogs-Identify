```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df1 = pd.read_csv('loss_resnet34.csv', header=None, skiprows=1)
df2 = pd.read_csv('loss_resnet50.csv', header=None, skiprows=1)
df3 = pd.read_csv('valid_los_resnet34.csv', header=None, skiprows=1)
df4 = pd.read_csv('valid_los_resnet50.csv', header=None, skiprows=1)

# 设置列名
df1.columns = ['index', 'value']
df2.columns = ['index', 'value']
df3.columns = ['index', 'value']
df4.columns = ['index', 'value']
x1 = np.linspace(1, 10, num=len(df1['value']))

# 绘制图表
plt.plot(x1, df1['value'], color='#16c19a', label='loss_resnet34')
plt.plot(x1, df2['value'], color='#005d2f', label='loss_resnet50')
plt.plot(df3['index'], df3['value'], '--', color='#ffa21d', label='valid_los_resnet34')
plt.plot(df4['index'], df4['value'], '--', color='#d07b00', label='valid_los_resnet50')

plt.xlabel('epochs')
plt.ylabel('Value')
plt.title('Loss')
plt.grid()
plt.legend()
plt.show()
```
