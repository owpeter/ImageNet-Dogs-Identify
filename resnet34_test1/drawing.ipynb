```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
data_dir = 'D:\Code\python\dog breed\loss'
nums = 7
loss = []
valid_loss = []

for i in range(nums):
    fname = 'loss_resnet34_no_pretrained_' + str(i) + '.csv'
    path = os.path.join(data_dir, fname)
    df = pd.read_csv(path, header=None, skiprows=1)
    df.columns = ['index', 'value']
    loss += list(df['value'])

for i in range(nums):
    fname = 'valid_los_resnet34_no_pretrained_' + str(i) + '.csv'
    path = os.path.join(data_dir, fname)
    df = pd.read_csv(path, header=None, skiprows=1)
    df.columns = ['index', 'value']
    valid_loss += list(df['value'])

x_loss = np.linspace(1, nums * 20, num=len(loss))
x_valid_loss = np.linspace(1, nums * 20, num=len(valid_loss))

plt.plot(x_loss, loss, label='loss')
plt.plot(x_valid_loss, valid_loss, label='valid_loss')

# # 绘制图表
# plt.plot(x1, df1['value'], color='#16c19a', label='loss_resnet34')
# plt.plot(x1, df2['value'], color='#005d2f', label='loss_resnet50')
# plt.plot(df3['index'], df3['value'], '--', color='#ffa21d', label='valid_los_resnet34')
# plt.plot(df4['index'], df4['value'], '--', color='#d07b00', label='valid_los_resnet50')

plt.xlabel('epochs')
plt.ylabel('Value')
plt.title('Loss')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.grid()
plt.legend()
plt.show()
```
